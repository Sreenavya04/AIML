```python
import pandas as pd
import matplotlib.pyplot as plt
people=['kiran','tom','cherry']
age=[11,12,13]
height=[4,5,6]
weight=[40,50,20]
data={
    'Name':people,
    'Age':age,
    'Height':height,
    'Weight':weight
}
df=pd.DataFrame(data)
print(df)
#SCATTER PLOT
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Weight'],color='blue',marker='o')
plt.title('Scatter plot of Age vs Weight')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kg)')
plt.figure(figsize=(8,6))
plt.hist(df['Age'],bins=5,color='orange',edgecolor='black')
plt.title('Histagram of Age Distribution')
plt.xlabel('Age(years)')
#HISTAGRAM
plt.figure(figsize=(8,6))
plt.bar(df['Name'],df['Height'],color='green')
plt.title('Bar char of individuals Height')
plt.xlabel('Name')
plt.ylabel('Height(cm)')
plt.show()
```
